-  How to extract all product link of a page
-  How to go through each link and extract Product Information
-  Handling Mismatched length of data (In some column some data is not avail.) with TRY/EXCEPT

Steps:
-  Import
-  Create Driver
-  Create Container ( To avoid getting upsell data and all)
-  Generate all 'href' links  (1 page or all)
-  Pass all link one by one in driver.get(link)
-  Extract Prdct Info
-  Create DataFrame

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import Select
import pandas as pd
import time
from tqdm import tqdm #Show progress in iterable code in %
```

```python
#Giving 'webdriver path location'
path= r"C:\Users\awadh\Desktop\WebScraping\chromedriver.exe"

#Popping up browser
s= Service(path)
driver=webdriver.Chrome(service=s)

#Giving website link
driver.get("https://www.amazon.com/s?k=Phone+under+20000&crid=3JAJO522J296L&sprefix=phone+under+1%2Caps%2C2286&ref=nb_sb_noss_2")
driver.maximize_window()
time.sleep(3)
```

```python
#CONTAINER
container=driver.find_element('xpath', "//span[@class='rush-component s-latency-cf-section']")
time.sleep(2)
#SINGLE PRODUCT
products=container.find_elements('xpath', "//div[contains(@class,'s-card-container')]")
```

```python
#Generate Link
links=[]

for i in tqdm(products):
    h2=i.find_element(By.TAG_NAME, "h2")
    anchor=h2.find_element(By.TAG_NAME, "a")
    href=anchor.get_property('href')
    links.append(href)  
    
#Extract Data
titles=[]
prices=[]
ratings=[]
reviews=[]

for link in tqdm(links):
    driver.get(link) #Passing individual link to driver
    time.sleep(4)
    
    #Extracting Data
    try:
        title=driver.find_element('xpath', '//div/h1/span').text
        titles.append(title)
    except:
        titles.append("")
    try:
        price=driver.find_element('xpath'," //td/span[contains(@class,'a-price a-text-price a-size-medium apexPriceToPay')]").text
        prices.append(price)
    except:
        prices.append("")
    try:
        rating=driver.find_element('xpath', '//*[@id="acrCustomerReviewText"][1]').text
        ratings.append(rating)
    except:
        ratings.append("")
    try:
        review=driver.find_element('xpath', "//div[@id='averageCustomerReviews']/span/span/span/a/span[1]").text
        reviews.append(review)
    except:
        reviews.append("")
        
    df= pd.DataFrame({'title':titles, 'price':prices, 'rating':ratings, 'review':reviews})
```
