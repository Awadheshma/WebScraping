```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
```

```python
names=[]
prices=[]
descriptions=[]
ratings=[]
reviews=[]
```

```python
for i in range(1,11):

    url='https://www.flipkart.com/search?q=Iphone&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page='+str(1)
    #checking request connection
    r= requests.get(url)
    #creating Soup
    soup= BeautifulSoup(r.text, 'lxml')
    #To avoid mismatch of len of data while creating df at last. BCoz it also extract data from below upsell product. 
    main_box= soup.find('div',class_='_1YokD2 _3Mn1Gg') 
    
    
    
    #name
    name= main_box.find_all('div', class_='_4rR01T')

    for i in name:
        a=i.text
        names.append(a)

    #price
    price=main_box.find_all('div', class_='_30jeq3 _1_WHN1')

    for i in price:
        b= i.text
        prices.append(b)

    #description
    desc= main_box.find_all('ul', class_='_1xgFaf')

    for i in desc:
        c=i.text
        descriptions.append(c)

    #rating
    rating= main_box.find_all('div', class_='_3LWZlK')

    for i in rating:
        d=i.text
        ratings.append(d)

    #review
    review= main_box.find_all('span',string=re.compile('Reviews'))  #No Class in html so extracting with help of Regular Expression

    for i in review:
        e=i.text
        reviews.append(e)
```

```python
#Checking Length of All columns:

# print(len(names))
# print(len(prices))
# print(len(descriptions))
# print(len(ratings))
# print(len(reviews))
```

```python
# Create Dataframe
df= pd.DataFrame({'Product_name':names, "Price":prices, "Rating":ratings, "Reviews":reviews,
                             'Description':descriptions})
```
